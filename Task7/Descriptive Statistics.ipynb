<h2>3. Mode</h2>
<p> Mode is defined as the value that is repeatedly occurring in a given set.</p>

### Formula of Mode

## $$
M_0 =\ l+h\left(\frac{f_1 -f_0}{2f_1-f_0-f_2}\right)\\
$$
Where,
<br> *l* = lower limit of the modal class,
<br> *h* = size of the class interval (assuming all class size to be equal),
<br> *f<sub>1</sub>* = frequency of the modal class,
<br> *f<sub>0</sub>* = frequency of the class preceding the modal class,
<br> *f<sub>2</sub>* = frequency of the class succeding the modal class.




```python
## taking the same data of annual_HH_income
from collections import Counter
c = Counter(annual_HH_income.values)
mode =  [k for k, v in c.items() if v == c.most_common(1)[0][1]]
print("Mode of annual_HH_income dataset : ",mode)
```

```python
plt.figure(figsize =(12,6))
graph =sns.scatterplot(y=annual_HH_income , x = range(len(annual_HH_income.values)), label = "Sample data points", marker = 'o',)
graph.axhline(median, label="Median", linestyle ="--")
plt.text(0.5,median+1000, f'Median ={median}', size ='14')
plt.legend(loc="upper left", fontsize = 20 )
graph.axhline(mean, label="Mean",color ="green", linestyle =":")
plt.text(0.5, mean+500, f'Mean = {mean}', size ='14')
graph.axhline(mode[0], label="Mode", color ="red")
plt.text(0.5, mode[0]+2000, f'Mode = {mode}', size ='14')

plt.title('Axhline() Example')
plt.legend(loc ='upper left')
```

### 4.Variance
- its the average squared deviation of the list elements from the average value.

<p>1. Population Variance,</p>
$$s^{2} = \frac{SS}{N} = \frac{\sum_{i=1}^{n} (x_{i} - \bar{x})^{2}}{N}$$

<p>2. Sample Variance </p>
$$s^{2} = \frac{SS}{N - 1} = \frac{\sum_{i=1}^{n} (x_{i} - \bar{x})^{2}}{N - 1}$$

### where,
<br>**SS**  = sum of squared errors</br>
<br>**N** = number of observations in the group</br>
<br>**x<sub>i</sub>** = **i<sup>th</sup>** score in a group</br> 
<br>$\mathbf{\bar{x}}$ = mean of the group.</br>
<br><b> Note: The Sample mean is one possible for the true population mean.</b></br>



```python
# 2. W/O External Dependency
lst= dataframe['Annual_HH_Income']
avg = sum(lst) / len(lst) # mean of gorup
var = sum((x-avg)**2 for x in lst) / (len(lst)-1) 
print("Variance: {}".format(var))
```

## $$s = \sqrt{s^{2}} = \sqrt{\frac{SS}{N - 1}} = \sqrt{\frac{\sum (x_{i} - \bar{x})^{2}}{N - 1}}$$

where,
<br>*S<sup>2</sup>**is the variance
<br>**SS**  is the sum of squared errors
<br>**N** is the number of observations in the group 
<br>**x<sub>i</sub>** is the **i<sup>th</sup>** score in a group; and 
<br>**$\bar{x}$** is the mean of the group.

```python
def get_std_dev(ls):
    n = len(ls)
    mean = sum(ls) / n
    var = sum((x - mean)**2 for x in ls) / n
    std_dev = var ** 0.5
    return std_dev
# create a list of data points
ls =dataframe['Annual_HH_Income'] 
standard_daviation = get_std_dev(ls)
print(standard_daviation)
```

- A correlation coefficient of 1 means that for every positive increase in one variable, there is a positive increase of a fixed proportion in the other. For example, shoe sizes go up in (almost) perfect correlation with foot length.
- A correlation coefficient of -1 means that for every positive increase in one variable, there is a negative decrease of a fixed proportion in the other. For example, the amount of gas in a tank decreases in (almost) perfect correlation with speed.
- Zero means that for every increase, there isnt a positive or negative increase. The two just arent related.

<h3>Correlation - Pearson</h3>
<p>The Pearson correlation coefficient is probably the most widely used measure for linear relationships between two normal distributed variables and thus often just called "correlation coefficient". Usually, the Pearson coefficient is obtained via a Least-Squares fit and a value of 1 represents a perfect positive relation-ship, -1 a perfect negative relationship, and 0 indicates the absence of a relationship between variables.</p>

### $$\rho = \frac{\text{cov}(X,Y)}{\sigma_x \sigma_y}$$

And the estimate

### $$r = \frac{{}\sum_{i=1}^{n} (x_i - \overline{x})(y_i - \overline{y})}
{\sqrt{\sum_{i=1}^{n} (x_i - \overline{x})^2\sum_{i=1}^{n}(y_i - \overline{y})^2}}$$

```python
def corr(X,Y):
    import numpy as np
    sum_diff =0
    x_sum, y_sum = 0,0
    for i in range(len(X)):
        x_bar = np.mean(X)
        y_bar = np.mean(Y)
        diff = (X[i]-x_bar)*(Y[i]-y_bar)
        sum_diff = diff+sum_diff
        x_sum = ((X[i]-x_bar)**2)+x_sum
        y_sum = ((Y[i]-y_bar)**2)+y_sum
    correlation = sum_diff/(np.sqrt(x_sum*y_sum))
    return correlation

X = dataframe.Emi_or_Rent_Amt
Y = dataframe.No_of_Earning_Members
```

<p><b>1. Normal distribution (univariate)</b></p>
<ul>
    
A random variable X is normally distributed with mean  and variance  if it has the probability density function of X as
<br><b>Note: Single Variable</b> <br>
    
 

##### Probability density function

#### $$ \phi(\pmb x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp^{ \bigg[-\frac{1}{2}\bigg( \frac{x-\mu}{\sigma}\bigg)^2 \bigg] } $$


<li>This result is the usual bell-shaped curve that you see throughout statistics.</li> 
<li>squared difference between the variable x and its mean $\mu$, $(x-\mu)^2$ .</li> 
$\phi(\pmb x)$ value will be minimized when x is equal to $\mu$. The quantity $ -\bigg( \frac{x-\mu}{\sigma}\bigg)^2 $ will take its largest value when x is equal to $\mu$ or likewise, since the exponential function is a monotone function, the normal density takes a maximum value when x is equal to $\mu$.
<p>The variance $\sigma ^2$ defines the spread of the distribution about that maximum. If $\sigma ^2$ is large, then the spread is going to be large, otherwise, if the $\sigma ^2$ value is small, then the spread will be small.</p>

As shorthand notation we may use the expression below:

#### $$ \phi(\pmb x) \sim N(\mu|\sigma^2) \\\\$$

### 8. Positively Skewed & Negatively Skewed Normal Distribution

- What is a Positively Skewed Distribution?
In statistics, a positively skewed (or right-skewed) distribution is a type of distribution in which most values are clustered around the left tail of the distribution while the right tail of the distribution is longer. The positively skewed distribution is a direct opposite of the negatively skewed distribution.

<b>Central Tendency Measures in Positively Skewed Distributions</b>
- Normally distributed data where all measures of the central tendency (mean, median, and mode) equal each other, with positively skewed data, the measures are dispersed.
- The general relationship among the central tendency measures in a positively skewed distribution may be expressed using the following inequality:
<br>$$Mean  >  Median  >  Mode $$</br>

- <h3>What is a Negatively Skewed Distribution?</h3>
In statistics, a negatively skewed (also known as left-skewed) distribution is a type of distribution in which more values are concentrated on the right side (tail) of the distribution graph while the left tail of the distribution graph is longer.
